```python
#WE ITERATE OVER OUR VENUES AND SEARCH FOR MATCHING LAT/LNG COUPLES
##WE NEED TO TRUNCATE TO ROUND UP, SO WE USE A FUNCTion, because the coords 
def truncate(f, n):
    '''Truncates/pads a float f to n decimal places without rounding'''
    s = '{}'.format(f)
    if 'e' in s or 'E' in s:
        return '{0:.{1}f}'.format(f, n)
    i, p, d = s.partition('.')
    return '.'.join([i, (d+'0'*n)[:n]])


#first we build a livenation coordtable
posLiveN = []
for data in dataLN:
    #print dataLN[data]["title"]
    posLiveN.append([dataLN[data]["title"],dataLN[data]["position"]])
    print [dataLN[data]["title"],dataLN[data]["position"]]
print posLiveN[0][1]["lat"]


#then we connect to the db and check venues for matching coord
#for venue in db.
nb_venues_total=0
nb_venues_LVN=0
count_total_shows=0
count_LVN_shows=0
print ("bandsintown coord lat/lng |  name of the venue | LiveNAtion coords lat/lng")
for venue in db.venues.find():
    nb_venues_total +=1
    count_total_shows    += int(venue["count"])
    for posLive in posLiveN:
        try:
            k=3
            if (truncate(float(posLive[1]["lat"]),k)== truncate(float(venue["latitude"]),k) and truncate(float(posLive[1]["lng"]),k)== truncate(float(venue["longitude"]),k)):
                print [venue["latitude"],venue["longitude"]], venue["name"] , [posLive[1]["lat"],posLive[1]["lng"]] , venue["count"], posLive[0]
                count_LVN_shows += int(venue["count"])
                nb_venues_LVN +=1
        except TypeError:
            print "TRR",[venue["latitude"],venue["longitude"]], venue["name"] 
print "nb_venues_total",nb_venues_total,"count_total",count_total_shows,"nb_venues_LVN",nb_venues_LVN,"count_LVN_shows",count_LVN_shows
```
