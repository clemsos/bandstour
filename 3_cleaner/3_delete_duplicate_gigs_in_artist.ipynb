```python
# -*- coding: utf-8 -*-
import re
def slugify(s):
    """
    Simplifies ugly strings into something URL-friendly.
    >>> print slugify("[Some] _ Article's Title--")
    some-articles-title
    """
    print s
    print type(s)
    s = re.sub('\s+', ' ', s)
    s = s.replace(' ', '-')

    s = re.sub('\W+', ' ', s)

    s = s.strip()

    s = s.lstrip()
    s = re.sub('\W+', '_', s)
    return s




filt=["more","etc", "and","\+","&","tba"]
#filt=["more "]
TOTCOUNTER=0
CSVDATA=[]
for fi in filt:
    print fi
    CSVDATA.append(["\""+fi+"\""])
    CSVDATA.append(["'"+str(fi)+"'"])
    regex = ".*" + fi + ".*";
    res = artists.find({"_id": {"$regex": regex, "$options": '-i'}})
    print "NB OCCURENCES:",res.explain()['executionStats']["nReturned"]
    COUNTER=0
    for re in res:
        #print re['_id']
        #print  
        if (fuzz.ratio(re['_id'].lower(), fi) >= 30) :
            print re['_id']
            CSVDATA.append(["\""+re['_id']+"\""])
            #CSVDATA.append([re['_id']])
            COUNTER+=1
    print COUNTER
    CSVDATA.append([])
    CSVDATA.append(["'"+str(COUNTER)+"'"])
    TOTCOUNTER+=COUNTER
    
print TOTCOUNTER
CSVDATA.append(["____"])
CSVDATA.append(["'"+str(TOTCOUNTER)+"'"])
print CSVDATA


myFile = open('etcTBAMOREKILLER.csv', 'w')
with myFile:
    writer = csv.writer(myFile)
    for row in CSVDATA:
        row=[r.encode('utf-8') for r in row]        
        writer.writerows([row])
     
print("Writing complete")
```
