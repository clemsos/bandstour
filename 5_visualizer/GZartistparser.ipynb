```python
# coding: utf8
Liste_GZ=[]
import re
from slugify import slugify
import pymongo

from pymongo import MongoClient

from fuzzywuzzy import fuzz
from fuzzywuzzy import process
client = MongoClient()
db = client["bandstour"]
venues =  db["venues"]
artists =  db["minedArtists"]
FUZZ_RATIO = 97 
```

```python
file = open("TITLES.TXT",'r')
text=file.readlines()
for a in text :
    ligne= a.strip("\nnn").strip('\"')
    #print ligne
    c= re.findall(r'(?<=:).*',ligne)#,re.MULTILINE| re.DOTALL | re.LOCALE )
    #print c
    #print len(c)
    if len(c)>0:
        #print len(c[0]) 
        if len(c[0]) >= 3 :
            d= re.split("\+|/|,|avec", c[0])
            #print d
            for lem in d:
                if "@" in lem:
                    lem2=lem.split("@")[0]
                else:
                    lem2=lem
                #print lem2
                Liste_GZ.append(slugify(lem2))
print Liste_GZ
print len(Liste_GZ)
```

```python
Liste_ART=[]
for artist in artists.find({},no_cursor_timeout=True) :
    Liste_ART.append(artist["slug"])

    
```

```python
Liste_GZ=set(Liste_GZ)
LIste_FOUNDGZ=[]
for band in Liste_GZ:
    for ART in Liste_ART:
        
        if (fuzz.ratio(band, ART)> FUZZ_RATIO):
            print band, ART
            LIste_FOUNDGZ.append(ART)
            
    
```
