```python

# coding: utf8
Liste_GZ=[]
import re
from slugify import slugify

file = open("TITLES.TXT",'r')
text=file.readlines()
for a in text :
    ligne= a.strip("\nnn").strip('\"')
    #print ligne
    c= re.findall(r'(?<=:).*',ligne)#,re.MULTILINE| re.DOTALL | re.LOCALE )
    #print c
    #print len(c)
    if len(c)>0:
        print len(c[0]) 
        if len(c[0]) >= 3 :
            d= re.split("\+|/|,|avec", c[0])
            #print d
            for lem in d:
                if "@" in lem:
                    lem2=lem.split("@")[0]
                else:
                    lem2=lem
                #print lem2
                Liste_GZ.append(slugify(lem2))
print Liste_GZ
```
